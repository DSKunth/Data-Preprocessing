## Dogs on Twitter

Dogs are also social media superstars! If we try to search "dog" in the people category on Twitter, a bunch of Twitter users will appear like below.







Dogs on Twitter are not just about bragging adorable dogs, some Twitter accountss are about guide dogs (like the one providing life-changing support, skills and dogs to people living with sight loss), some are related to dog welfare charities, and some are professional dog raters.





## WeRateDogs

The top on search result is Twitter user @dog_rates also known as WeRateDogs. According to their Twitter Bio, they are *'Your Only Source For Professional Dog Ratings.'* WeRateDogs joined Twitter in November 2015 and has 9.2M followers and 15.3K Tweets as of today. They are averaging about 6 Tweets per day. The account was started in 2015 by college student Matt Nelson, and has received international media attention for its popularity.







WeRateDog has a *Dog Submission Portal* [WeRateDogs](https://links.weratedogs.com/) where followers can submit photos (or videos) of their dogs with some 'tell me something about your dog'. WeRateDogs then tweets selected photos with rating and humorous comment. According to [WeRateDog Wikipedia](https://en.wikipedia.org/wiki/WeRateDogs), they receive 800 to 1,000 submissions daily.








## WeRateDogs Tweets

WeRateDogs are committed to humorously reviewing pictures of dogs doing adorable poses, often giving them hilarious rating score. Dogs are rated on a scale of one to ten, but are invariably given ratings in excess of the maximum, such as "14/10". 



And check this out - with rating of 144/120 - because they are 12 dogs!




Aside from amusing comments and crazy ratings, some tweets state the name, the breed and the stage of dogs. Yes, dog stage. There is a so-called Dogtionary [WeRateDogs book on Amazon](https://www.amazon.com/WeRateDogs-Most-Hilarious-Adorable-Youve/dp/1510717145) that explains the various stages of dog: doggo, pupper, puppo, and floof(er).







## Insights
We have gathered WeRateDogs twitter archive data from their first ever tweet on November 15, 2015 until August 1, 2017. We pre-processed the data and turned it into interesting insights.





## Twitter for iPhone is the utility mostly used by WeRateDogs for posting tweets.



Pretty sure WeRateDogs is an iPhone user. Nelson uses Twitter for iPhone almost all the time as opposed to Vine, Web or TweetDeck.




## Is there a relationship between the Likes and Retweets associated with dog breeds?



Setting aside the 'unknown' which refers to those dogs whose breeds are not mentioned in the tweets, **Golden Retriever** took the tops spot with 586,657 Retweets and 1,943,782 Likes. Followed by **Labrador Retriever** with 405,404 Retweets and 1,265,355 Likes.  **Pembroke** comes third with 291,771 Retweets and 1,038,797 Likes.





## What about the relationship between Likes and Retweets associated with dog stages?



In this bar graph, pupper has the longest bars with total Retweets of 528,234 and total Likes of 1,543,210, followed by doggo which has total Retweets of 466,193 and total likes of 1,272,012.

What's a pupper? a doggo? According to the Dogtionary, *a pupper is a small doggo, usually younger, can be equally , if not more mature than some doggos!* While *a doggo is a big pupper , usually older, and this label does not stop a doggo from behaving like a pupper!*
